```python
import nsepy
import datetime
import pandas as pd
```

```python
def gethdata(stockname):
    today=datetime.date.today()
    duration=100
    duration2=44
    start=today+datetime.timedelta(-duration)
    end1=today+datetime.timedelta(-duration2)
    stockdata=nsepy.get_history(symbol=stockname,start=start,end=end1)
    return stockdata
rf=gethdata("HDFC")
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
```

```python
train= sd
train= train.dropna()
```

```python
lr = LinearRegression()
 
X_train = train[['Prev Close','Deliverable Volume','VWAP','Low','Trades','Open']]
Y_train = train['High']
 
z=lr.fit(X_train,Y_train)
```

```python
def gethdatatest(stockname):
    today=datetime.date.today()
    duration=44
    
    start=today+datetime.timedelta(-duration)
    
    stockdata=nsepy.get_history(symbol=stockname,start=start,end=today)
    return stockdata
df1=gethdatatest('HDFC')
```

```python
test=df1[['Prev Close','High','Deliverable Volume','VWAP','Low','Trades','Open']]
```

```python
test.head()
```

```python
testx=test[['Prev Close','Deliverable Volume','VWAP','Low','Trades','Open']]
q= z.predict(testx)

```

```python
test['high prediction']=q
```

```python
test
```

```python
a=[]
i=0
t=test.index.values
for dt in t:
    a.append(dt.strftime("%m/%d/%Y"))
fig, ax = plt.subplots(figsize=(20, 20))
xd = [datetime.datetime.strptime(d,"%m/%d/%Y").date() for d in a]

ax.plot(xd, test['High'],'-o',label = "High", color = 'red')
ax.bar(xd, test['high prediction'], label = "predicted high by the model", color = 'green')


ax.grid()
plt.legend()
plt.show()
```
